# 01 — Loading & Overview

This notebook loads five years of DVF data (2021–2025) from `data/data_csv/`, concatenates the files, and provides a first overview of the raw dataset.

```python
from pathlib import Path
import pandas as pd

DATA_DIR = Path('data/data_csv')
years = range(2021, 2026)
files = [DATA_DIR / f'ValeursFoncieres-{y}.csv' for y in years]

frames = []
for fp in files:
    tmp = pd.read_csv(fp, sep='|', decimal=',', low_memory=False)
    tmp.columns = tmp.columns.str.strip()
    frames.append(tmp)
    print(f'  {fp.name}: {len(tmp):,} rows')

df = pd.concat(frames, ignore_index=True)
print(f'\nTotal rows after concat: {len(df):,}')
df.head()
```

```python
print('shape:', df.shape)
print()
print('columns:', len(df.columns))
print('missing values (top 20):')
print(df.isna().sum().sort_values(ascending=False).head(20))
```
